# Exploratory Data Analysis 
## Customer Analysis 

### Customer segmentation 

- Group customers based on segments 

```python
# Types of customers

types_of_customers = df['Segment'].unique()
print(types_of_customers)
```

```python
# Number of customers in each segment 

customer_segmentation = df['Segment'].value_counts().reset_index() 
customer_segmentation = customer_segmentation.rename(columns={'index': 'Customer Type', 'Segment': 'Total Customers'})

print(customer_segmentation)
```

```python
# Plotting the pie chart
plt.figure(figsize=(6, 6))  # Adjust figure size as needed
plt.pie(customer_segmentation['Total Customers'], labels=customer_segmentation['Customer Type'], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Centering the pie chart
plt.gca().set_aspect('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adding title with customized font properties
plt.title("Distribution of Customers", fontsize=16, color='blue', fontweight='bold')  # Adjust font size, color, and weight as needed

# Displaying the plot
plt.show()
```

Above chart shows the customer analysis i.e there are almost `52% consumers`, `30% Corporate`, `18% Home Office`.

```python
# Customers and Sales 

sales_per_segment = df.groupby('Segment')['Sales'].sum().reset_index()
sales_per_segment = sales_per_segment.rename(columns={'Segment': 'Customer Type', 'Sales': 'Total Sales'})

print(sales_per_segment)
```

```python
plt.figure(figsize=(6, 6))
plt.bar(sales_per_segment['Customer Type'], sales_per_segment['Total Sales'], color=['skyblue', 'lightgreen', 'salmon'])

# Title and Label
plt.title('Sales by Customer Type', fontsize=16)
plt.xlabel('Customer Type', fontsize=14)  
plt.ylabel('Sales', fontsize=14) 

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

The chart above displays sales by customer type, indicating that the `consumer` category dominates more than "corporate" and "home office" categories.

### Customer Loyalty
- Examine the repeat purchase behavior of customers 

```python
# Group the data by Customer ID, Customer Name, Segments, and calculate the frequency of orders for each customer
customer_order_frequency = df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Order ID'].count().reset_index()

# Rename the column to represent the frequency of orders
customer_order_frequency.rename(columns={'Order ID': 'Total Orders'}, inplace=True)

# Identify repeat customers (customers with order frequency greater than 1)
repeat_customers = customer_order_frequency[customer_order_frequency['Total Orders'] >= 1]

# Sort "repeat_customers" in descending order based on the "Order Frequency" column
repeat_customers_sorted = repeat_customers.sort_values(by='Total Orders', ascending=False)

# Print the result- the first 10 and reset index
print(repeat_customers_sorted.head(12).reset_index(drop=True))
```

This code essentially helps to identify and rank repeat customers based on their order frequency, providing insights into customer behavior and loyalty.

### Sales by Customer 
- Identify top-spending customers based on their total purchase amount 

```python
# Group the data by customer IDs and calculate the total purchase (sales) for each customer
customer_sales = df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Sales'].sum().reset_index()

# Sort the customers based on their total purchase in descending order to identify top spenders
top_spenders = customer_sales.sort_values(by='Sales', ascending=False)

# Print the top-spending customers
print(top_spenders.head(10).reset_index(drop=True))
```

Above output helps to identify and rank customers based on their total spending, providing insights into the most valuable customers in terms of sales volume.

### Shipping 

```python
# Types of Shipping methods 

types_of_customers = df['Ship Mode'].unique()
print(types_of_customers)
```

```python
# Frequency of use of a shipping methods 

shipping_model = df['Ship Mode'].value_counts().reset_index()
shipping_model = shipping_model.rename(columns={'index':'Mode_of_Shipment', 'Ship Mode': 'Use Frequency'})

print(shipping_model)
```

```python
# Plotting the pie chart
plt.figure(figsize=(6, 6))  # Adjust figure size as needed
plt.pie(shipping_model['Use Frequency'], labels=shipping_model['Mode_of_Shipment'], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Centering the pie chart
plt.gca().set_aspect('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adding title with customized font properties
plt.title("Popular Mode of Shipment", fontsize=16, color='blue', fontweight='bold')  # Adjust font size, color, and weight as needed

# Displaying the plot
plt.show()

```

The chart above illustrates the most popular shipping modes among customers, with approximately `60% preferring standard class`, `20% opting for second class`, `15% choosing first class`, and `6% selecting same day delivery`.

```python
# Plotting the pie chart
plt.figure(figsize=(6, 6))  # Adjusting figure size for better visualization
plt.pie(top_product_category['Sales'], labels=top_product_category['Category'], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Adding a title
plt.title('Top Product Categories Based on Sales', fontsize=16, fontweight='bold')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
```

```python
# Group the data by product sub category versus the sales  
product_subcategory = df.groupby(['Sub-Category'])['Sales'].sum().reset_index()

# Sort the product category in their descending order and identify top product category
top_product_subcategory = product_subcategory.sort_values(by='Sales', ascending=False) 

# Print the states 
print(top_product_subcategory.reset_index(drop=True))
```

```python
import matplotlib.pyplot as plt

# Sorting the DataFrame by sales in ascending order for better visualization
top_product_subcategory = top_product_subcategory.sort_values(by='Sales', ascending=True)

# Plotting the horizontal bar graph
plt.figure(figsize=(10, 6))  # Adjusting figure size for better visualization
bars = plt.barh(top_product_subcategory['Sub-Category'], top_product_subcategory['Sales'], color='skyblue')

# Adding data labels on the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, '{:.2f}'.format(bar.get_width()), 
             va='center', ha='left', fontsize=8)

# Adding title and labels
plt.title('Top Product Sub-Categories Based on Sales', fontsize=16, fontweight='bold')
plt.xlabel('Total Sales', fontsize=12)
plt.ylabel('Product Sub-Category', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Adding horizontal grid lines for better readability

# Show plot
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```

The above chart shows the 'Top Products Categroy Based On Sales' which illustrates that phones amd chairs are the top sold products followed by other items.

## Sales 

```python
# Convert the "Order Date" column to datetime format

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Group the data by years and calculate the total sales amount for each year
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()

yearly_sales = yearly_sales.reset_index()
yearly_sales = yearly_sales.rename(columns={'Order Date': 'Year', 'Sales':'Total Sales'})

# yearly_sales = 
# Print the total sales for each year
print(yearly_sales)
```

```python
import matplotlib.pyplot as plt

# Plotting the bar graph
plt.figure(figsize=(10, 6))  # Adjusting figure size for better visualization
bars = plt.bar(yearly_sales['Year'], yearly_sales['Total Sales'], color='skyblue')

# Adding data labels on the bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), '{:,.0f}'.format(bar.get_height()), 
             ha='center', va='bottom', fontsize=9)

# Adding title and labels
plt.title('Yearly Sales', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding vertical grid lines for better readability

# Show plot
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a line graph
plt.figure(figsize=(8, 6))  # Adjusting figure size for better visualization
plt.plot(yearly_sales['Year'], yearly_sales['Total Sales'], marker='o', linestyle='-', color='skyblue', linewidth=2)

# Adding data labels at each data point
for x, y in zip(yearly_sales['Year'], yearly_sales['Total Sales']):
    plt.text(x, y, '{:,.0f}'.format(y), ha='right', va='bottom', fontsize=9)

# Adding title and labels
plt.title('Total Sales by Year', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(yearly_sales['Year'])  # Set x-axis ticks to match years
plt.grid(axis='both', linestyle='--', alpha=0.7)  # Adding grid lines for better readability

# Show plot
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```

## Sales Trends

```python
import matplotlib.pyplot as plt

# Convert the "Order Date" column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Group the data by months and calculate the total sales amount for each month
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

# Plot the sales trends for months
plt.figure(figsize=(12, 6))  # Adjusting figure size for better visualization

# Monthly Sales Trend
plt.plot(monthly_sales.index.to_timestamp(), monthly_sales, marker='o', linestyle='-')  # Convert period index to timestamp for plotting
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales Amount', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding horizontal grid lines for better readability
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show plot
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```
